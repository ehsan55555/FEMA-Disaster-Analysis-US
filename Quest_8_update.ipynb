```python

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the data from the CSV file
data = pd.read_csv('us_disaster_declarations.csv')

num_declarations = len(data)

state_counts = data['state'].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(state_counts.index, state_counts.values, color='blue')
plt.xlabel('State')
plt.ylabel('Number of Disaster Declarations')
plt.title('Distribution of Natural Disaster Declarations Across US States')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Step 5: Print the total number of declarations
print("Total Number of Natural Disaster Declarations in the Dataset:", num_declarations)
```
