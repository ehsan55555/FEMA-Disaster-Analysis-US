```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the natural_disasters.csv file
df = pd.read_csv('us_disaster_declarations.csv')

df['declaration_date'] = pd.to_datetime(df['declaration_date'])

df['year'] = df['declaration_date'].dt.year

disaster_frequency = df.groupby(['year', 'incident_type']).size().reset_index(name='frequency')

severity_attributes = ['declaration_title']
df['severity_index'] = df[severity_attributes].sum(axis=1)
disaster_severity = df.groupby(['year', 'incident_type'])['severity_index'].sum().reset_index()

print(disaster_severity)
print(severity_attributes)

 #Define a color index for each natural disaster
disaster_colors = {
    'Fire': 'red',
    'Flood': 'blue',
    'Other': 'green',
    'Tornado': 'purple',
    'Earthquake': 'orange',
    'Snowstorm': 'cyan',
    'Hurricane': 'yellow',
    'Volcanic Eruption': 'brown',
    'Severe Storm': 'coral',
    'Toxic Substances': 'crimson',
    'Typhoon': 'darkblue',
    'Dam/Levee Break': 'darkgreen',
    'Drought': 'gold',
    'Sever Ice Storm': 'grey',
    'Freezing': 'ivory',
    'Costal Storm': 'indigo',
    'Fishing Losses': 'khaki',
    'Mud/Landslide': 'lavender',
    'Human Cause': 'lightblue',
    'Terroist': 'lightgreen',
    'Chemical': 'lime',
    'Biological': 'olive',
    'Tropical Storm': 'pink',
    'Winter Storm': 'salmon'}


# Create a bar graph for the frequency of disasters over the years
plt.figure(figsize=(13, 8))
for disaster in disaster_frequency['incident_type'].unique():
    color = disaster_colors.get(disaster)
    plt.bar(disaster_frequency[disaster_frequency['incident_type'] == disaster]['year'],
            disaster_frequency[disaster_frequency['incident_type'] == disaster]['frequency'],
            label=disaster, color=color)

plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Frequency of Disasters Over the Years')

handles, labels = plt.gca().get_legend_handles_labels()
for label, handle in zip(labels, handles):
    handle.set_label(label + ' (Color Index: ' + disaster_colors.get(label, 'gray') + ')')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# Define a color index for each natural disaster
disaster_colors = {
    'Fire': 'red',
    'Flood': 'blue',
    'Other': 'green',
    'Tornado': 'purple',
    'Earthquake': 'orange',
    'Snowstorm': 'cyan',
    'Hurricane': 'yellow',
    'Volcanic Eruption': 'brown',
    'Severe Storm': 'coral',
    'Toxic Substances': 'crimson',
    'Typhoon': 'darkblue',
    'Dam/Levee Break': 'darkgreen',
    'Drought': 'gold',
    'Sever Ice Storm': 'grey',
    'Freezing': 'ivory',
    'Costal Storm': 'indigo',
    'Fishing Losses': 'khaki',
    'Mud/Landslide': 'lavender',
    'Human Cause': 'lightblue',
    'Terroist': 'lightgreen',
    'Chemical': 'lime',
    'Biological': 'olive',
    'Tropical Storm': 'pink',
    'Winter Storm': 'salmon'}

# Line plot for frequency of disasters over the years
plt.figure(figsize=(13, 8))
for disaster in disaster_frequency['incident_type'].unique():
    color = disaster_colors.get(disaster)
    plt.plot(disaster_frequency[disaster_frequency['incident_type'] == disaster]['year'],
             disaster_frequency[disaster_frequency['incident_type'] == disaster]['frequency'],
             label=disaster, color=color)
    
handles, labels = plt.gca().get_legend_handles_labels()
for label, handle in zip(labels, handles):
    handle.set_label(label + ' (Color Index: ' + disaster_colors.get(label, 'gray') + ')')

plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Frequency of Disasters Over the Years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

    
     


   


```
